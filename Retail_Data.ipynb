```python
import pandas as pd

# Path to the CSV file
file_path = '/content/drive/MyDrive/Colab Notebooks/new_retail_data.csv'

# Read the CSV file into a DataFrame
main_df = pd.read_csv(file_path)

# Display the first few rows
main_df.head()
```

```python
# Replace empty strings in 'Customer_ID' with NA (safer approach)
main_df['Customer_ID'] = main_df['Customer_ID'].replace('', pd.NA)

# Fill missing 'Customer_ID' using Email
main_df['Customer_ID'] = main_df['Customer_ID'].fillna(
    main_df['Email'].map(
        main_df[main_df['Customer_ID'].notna()].drop_duplicates('Email').set_index('Email')['Customer_ID']
    )
)
```

```python
# Get rows where Customer_ID is blank, NaN, or None
main_df[main_df['Customer_ID'].isna() | (main_df['Customer_ID'] == '')]
```

```python
# Replace empty strings in 'Email' with NA
main_df['Email'] = main_df['Email'].replace('', pd.NA)

# Fill missing 'Email' using Customer_ID
main_df['Email'] = main_df['Email'].fillna(
    main_df['Customer_ID'].map(
        main_df[main_df['Email'].notna()].drop_duplicates('Customer_ID').set_index('Customer_ID')['Email']
    )
)
```

```python
# Get rows where Email is blank, NaN, or None
main_df[main_df['Email'].isna() | (main_df['Email'] == '')]
```

```python
columns_to_fill = ['Name','Address', 'City', 'State', 'Country', 'Gender', 'Income', 'Phone', 'Zipcode']

for col in columns_to_fill:
    # Replace empty strings with NA
    main_df[col] = main_df[col].replace('', pd.NA)

    # Fill missing values using Customer_ID
    main_df[col] = main_df[col].fillna(
        main_df['Customer_ID'].map(
            main_df[main_df[col].notna()].drop_duplicates('Customer_ID').set_index('Customer_ID')[col]
        )
    )

    # Fill remaining missing values using Email
    main_df[col] = main_df[col].fillna(
        main_df['Email'].map(
            main_df[main_df[col].notna()].drop_duplicates('Email').set_index('Email')[col]
        )
    )
```
